```python
phonebook = {
                0: {'Name': 'Kim', 'Phone': '45648733'},
                1: {'Name': 'Lee', 'Phone': '89376333'},
                2: {'Name': 'Park', 'Phone': '36457818'},
                3: {'Name': 'Chung', 'Phone': '76891234'},
                4: {'Name': 'Choi', 'Phone': '67451237'}
            }



# JSON 파일을 읽고 쓰기 위해 json 모듈을 가져온다.
import json  

#전화번호부 목록 출력 함수
def list_phonebook():  # 전화번호부에 있는 모든 멤버 정보를 출력하는 함수
    """ List member in phone book. """ # """ 함수 바로 아래에 적으면 함수, 클래스, 모듈의 설명 문서가 됨."""
    
    f = open('../source/45-1.json', 'r')  # JSON 파일을 읽기 모드로 연다
    d = json.load(f)  # 파일 내용을 딕셔너리 형태로 읽어온다

    for pid in d:  # 딕셔너리의 key(pid)를 하나씩 꺼내면서 반복
        print('\nPID:', int(pid) + 1)  # pid는 문자열이므로 정수로 바꿔서 출력 (번호를 1부터 시작하게)
        for p_info in d[pid]:  # 한 사람 정보 안의 key(Name, Phone 등)를 반복
            print(p_info + ':', d[pid][p_info])  # key와 value를 함께 출력한다

    f.close()  # 파일 닫기


#새 멤버 추가 함수
def add_member():  # 전화번호부에 새 멤버를 추가하는 함수
    """ Add a new member to the phone book."""

    print("\nEnter the information of the member")  # 사용자에게 안내 메시지 출력

    name = input('Name: ')  # 이름 입력 받기
    phone = input('Phone: ')  # 전화번호 입력 받기

    name_check = False  # 중복 이름 확인용 = 중복된 이름 찾지 못함(초기이름)

    f = open('../source/45-1.json', 'r')  # JSON 파일을 읽기 모드로 연다
    d = json.load(f)  # 딕셔너리로 불러오기
    f.close()  # 파일 닫기

    for pid in d:  # 멤버 한 명씩 반복
        if name == d[pid].get('Name'):  # 이미 같은 이름이 있는지 확인
            name_check = True  # 있으면 True로 표시

    if name_check is True:  # 중복되면
        print('\n# The member is already in the phone book')  # 안내 메시지 출력
    else:
        with open('../source/45-1.json', 'w') as f:  # 쓰기 모드로 파일 열기 (덮어쓰기)
            d[len(d)] = {'Name': name, 'Phone': phone}  # 새 멤버를 딕셔너리에 추가

            json_obj = json.dumps(d, indent=4)  # 딕셔너리를 JSON 문자열로 변환 (들여쓰기 예쁘게)
            f.write(json_obj)  # 파일에 쓰기
            print('\n# The phone number has been written to the phone book')  # 완료 메시지 출력

    return d  # 딕셔너리 반환 (필수는 아님, 확인용)

#멤버 삭제 함수
def delete_member():  # 전화번호부에서 특정 멤버를 삭제하는 함수
    """ Delete a member from the phone book. """

    print("\nEnter the name")  # 사용자에게 안내
    name = input('name: ')  # 삭제할 이름 입력

    f = open('../source/45-1.json', 'r')  # 파일 열기 (읽기 모드)
    d = json.load(f)  # 딕셔너리로 불러오기
    f.close()  # 파일 닫기

    for pid in d:  # 멤버 하나씩 확인
        if name == d[pid].get('Name'):  # 이름이 일치하면
            with open('../source/45-1.json', 'w') as f:  # 쓰기 모드로 파일 열기, → 딕셔너리에서만 삭제하면 실제 파일에는 반영되지 않기 때문
                del d[pid]  # 해당 멤버 삭제 덮어쓰기
                json_obj = json.dumps(d, indent=4)  # 다시 JSON 문자열로 변환
                f.write(json_obj)  # 파일에 저장
                print('\n# The member has been deleted')  # 완료 메시지
                return d  # 삭제 후 딕셔너리 반환

    print('\n# The member is not in the phone book')  # 해당 이름이 없을 때 메시지

#메인 메뉴 (사용자 입력 받기)
def mainmenu():  # 메인 메뉴를 보여주고 사용자의 선택에 따라 작업을 실행하는 함수
    """ Main function"""

    while True:  # 무한 반복
        menu = input("""
----MAIN MENU----
1: List phonebook
2: Add a new member
3: Delete a member
4: Program exit 
Please enter your choice: """)  # 메뉴 안내 + 사용자 입력 받기

        if menu == '1':  # 목록 보기
            list_phonebook()
        elif menu == '2':  # 추가
            add_member()
        elif menu == '3':  # 삭제
            delete_member()
        elif menu == '4':  # 종료
            print('# Exit!')
            return False  # 반복문 종료
        else:
            print('\n Menu cannot be found')  # 잘못된 입력 안내

#프로그램 시작
mainmenu()  # 메인 메뉴 함수 실행 (프로그램 시작점)
```
