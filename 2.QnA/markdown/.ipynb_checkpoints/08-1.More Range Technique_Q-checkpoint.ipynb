````python
# 8. More Range Technique


#ë°©ë²• 1
ex1 = []
for i in range(1, 21):
    if i % 2 != 0:
        ex1.append(i * 10)
    else:
        ex1.append(i)

print(ex1)


#ë°©ë²• 2
z = [x * 10 if x % 2 != 0 else x for x in range(1, 21)]
print(z)


# ë¦¬ìŠ¤íŠ¸ì»´í”„ë ˆì´ì…˜(ë¦¬ìŠ¤íŠ¸ ì•ˆì— ë¦¬ìŠ¤íŠ¸ë¥¼ ë”°ë¡œ ë§Œë“¤ê¸°)
print([[j for j in range(5)] for i in range(5)])
````
